### 整体风格设置

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*.5)*(7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

### 6个默认的颜色循环主题：deep, muted, pastel, bright, dark, colorblind

### 圆形画板
    当你有6个以上的分类要区分时，最简单的方法就是在一个圆形的颜色空间中画出均匀间隔的颜色（这样的色调会保持亮度和饱和度不变）。这是大多数的当他们需要使用比当前默认颜色循环中设置的颜色更多时的默认方案。
    最常用的方法是使用hls的颜色空间，这是RGB值的一个简单转换。

```python
sns.palplot(sns.color_palette('hls', 8))
```

```python
sns.palplot(sns.color_palette('hls', 12))
```

```python
data = np.random.normal(size=(20, 8)) + np.arange(8) / 2
sns.boxplot(data=data, palette=sns.color_palette('hls', 8))
```

### 使用xkcd颜色来命名颜色
    xkcd包含了一套众包努力的针对随机RGB色的命名。产生了954个可以随时通过xkcd_rgb字典中调用的命名颜色。

```python
plt.plot([0, 1], [0, 1], sns.xkcd_rgb['light red'], lw=1)
plt.plot([0, 1], [0, 2], sns.xkcd_rgb['medium green'], lw=3)
plt.plot([0, 1], [0, 3], sns.xkcd_rgb['light blue'], lw=3)
```

### 单变量分析绘图

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
np.random.seed(sum(map(ord, 'distributions')))
```

```python
x = np.random.normal(size=100)
# plt.hist(x, density=False)
sns.distplot(x, kde=True)
```

```python
sns.distplot(x, bins=20, kde=False, rug=True)
```

### 数据分布情况

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma)
```

### 根据均值和协方差生成数据

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=['x', 'y'])
df.head()
```

 ### 观测两个变量之间的分布关系最好用散点图

```python
sns.jointplot(x='x', y='y', data=df)
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style('white'):
    sns.jointplot(x=x, y=y, kind='hex', color='k')
```
